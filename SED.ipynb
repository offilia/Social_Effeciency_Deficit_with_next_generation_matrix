```python
%%time
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import matplotlib.ticker as ticker

#次世代行列によるRoの計算
#arr = np.zeros((101, 101), dtype = float) #Roヒートマップとして保存する場合
#x1 = np.zeros(101, dtype = float)
#y1 = np.zeros(101, dtype = float)
x = []
y = []
Ro = 2.5
rs = 1 / 3
ra = 1 / 3
B = Ro * rs
S = 0.9995
ts = 0.4
ta = 0.1

#Ro境界線1.0、1.2
def Ro(N,start,end,step):
    for Q in range(start, end, step):#tauS,0~0.3:(100, -1, -1) or 0.4~:(0, 101)
        check = True
        for P in range(start, end, step):#(100, -1, -1) or (0, 101)
            q = Q * 0.01
            p = P * 0.01
            Ba = B * q
            #arr[Q][P] = (p*Ba)/(ta+rs) + B*(1-p)/(ts+rs)#Roヒートマップとして保存する場合
            arr = (p*Ba)/(ta+rs) + B*(1-p)/(ts+rs)
            if check:
                if arr > N * 0.1:
                    #x1[Q] = P #Roヒートマップとして保存する場合
                    #y1[Q] = Q
                    x.append(P)
                    y.append(Q)
                    check = False

Ro(10,100,-1,-1)
Ro(12,0,101,1)
                    
#Roのヒートマップ保存
#df = pd.DataFrame(arr)
#df.to_csv('new_tauS_0.1.csv')

#ASPneデータの加工
array_ne = np.zeros((101, 101), dtype = float)
filedir = input("ASPneのファイルのパスを入力してください")
filedir = filedir.replace('"','').strip()
dataset_2 = pd.read_csv(filedir, header=None)
dataset_2 = dataset_2.drop(dataset_2.index[0], axis=1)
dataset_2 = dataset_2.drop(dataset_2.index[0], axis=0)
for i in range(101):
    for j in range(101):
        array_ne[i][j] = dataset_2[i+1][j+1]

#ASPsoデータの加工
array_so = np.zeros((101, 101), dtype = float)
filedir = input("ASPsoのファイルのパスを入力してください")
filedir = filedir.replace('"','').strip()
dataset = pd.read_csv(filedir)
dataset.sort_values(['p', 'q'], inplace=True)
dataset.reset_index(inplace = True, drop=True)
for i in range(10201):
    print(dataset.asp[i])
    array_so[int(i/101)][int(i%101)] = dataset.asp[i]

#SEDの計算
array_sed = np.zeros((101, 101), dtype = float)
Colum = ['q','p','sed']
l=[]
for i in range(101):
    for j in range(101):
        array_sed[i][j] = array_so[j][i] - array_ne[j][i]#横軸p、縦軸qのデータ構造に書き換え
        l.append([i * 0.01, j * 0.01, array_sed[i][j]])
df = pd.DataFrame(l, columns = Colum)
print("SEDの最小値：", df['sed'].min(), "SEDの最大値：", df['sed'].max())#SEDの最小値と最大値の確認


#ヒートマップの作成!!

#SEDのファイル名を決める
filename = input("ファイル名なんにするん？？")
#データ読み込み
dataset = df
#データフレームの調整
dataset = dataset.pivot("q","p","sed")

#SED以外(xmax,ASPso,It,Vt)のヒートマップを作成する場合
#filedir = input("ファイルのパスを入力してください")
#filedir = filedir.replace('"','').strip()
#filename=os.path.basename(filedir)
#dataset = pd.read_csv(filedir)
#dataset = dataset.pivot("q","p","asp")

ax=sns.heatmap(data = dataset,
               cmap = "Greys",
               fmt = ".2f",
               vmax = 10, # df['sed'].max() + 1
               vmin = -1, # df['sed'].min() - 1
               robust = False, #True に設定しvminとvmaxが設定されていない場合、カラーマップが使う色とその値の範囲はロバストな推定を行い、極端に大きな値や小さい値の影響を受けないように計算されます。
               annot = False,
               annot_kws = None,
               #linewidths = 0, #各セル間の線の太さ(デフォルト値: 0)
               #linecolor = "Blue", #各セル間の線の色(デフォルト値: ‘white’)
               #center = dataset.loc[0.5, 0.5],#centerでセルを指定すると、そのセル部分を中央値の黒色としてヒートマップが形成される
               square = True,
               xticklabels = 50,
               yticklabels = 50,
               cbar = True,
               cbar_ax = None, #カラーバーを出力する際の軸を Matplotlib の軸で指定する
               #mask = mask,
               cbar_kws = {'shrink': 0.6})

fsize=30
ax.figure.axes[-1].set_ylabel('SED', size=25, rotation= 0, labelpad=50)
ax.figure.set_size_inches(10,10)
ax.invert_yaxis()
plt.title("",fontsize = fsize)#tauS=0.1,Cv=0.5,d=0.01
plt.xlabel("p",fontsize = fsize)
plt.xticks(fontsize = fsize-10, rotation = 0)
plt.ylabel("q",fontsize = fsize)
plt.yticks(fontsize = fsize-10, rotation = 0)
plt.plot(x, y, marker = "o", color = "red", linestyle='None')
plt.tight_layout() #サブプロット間の正しい間隔を自動的に維持してくれちゃう、これを使わない場合1つの行が次の行のタイトルと重なることがある、サブプロット間の間隔を変更⇒plt.subplots_adjust(left=0.125,bottom=0.1, right=0.9, top=0.9, wspace=0.2, hspace=0.35)
#出力結果保存
plt.savefig(filename +'.png', bbox_inches='tight')
```
